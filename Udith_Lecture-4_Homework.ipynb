```python
data_usa = data_frame_confirmed[data_frame_confirmed['Country/Region'] == 'US']
death_usa = data_frame_deaths[data_frame_deaths['Country/Region'] == 'US']
recover_usa = data_frame_recovered[data_frame_recovered['Country/Region'] == 'US']
#data_korea = data_korea.drop(columns=['Province/State', 'Country/Region', 'Intermediate Region Code'])
val_usa = data_usa['Value'].to_numpy().astype(int)
val2_usa = death_usa['Value'].to_numpy().astype(int)
val3_usa = recover_usa['Value'].to_numpy().astype(int)
val_usa = np.flipud(val_usa)
val2_usa = np.flipud(val2_usa)
val3_usa = np.flipud(val3_usa)
plt.plot(val_usa, label='Confirmed', color='blue')
plt.plot(val2_usa, label='Deaths', color='red')
plt.plot(val3_usa, label='Recovered', color='green')
legend = plt.legend(loc='upper left', shadow=True, fontsize='medium')
plt.xlabel("Days")
plt.ylabel("Cases")
plt.title("USA")
plt.show()
```

```python
#Computing the infection Rate 

infectionrate_usa = np.gradient(val_usa)
infectionrate_spain = np.gradient(val_spain)
infectionrate_italy = np.gradient(val_italy)
infectionrate_uk = np.gradient(val_uk)
infectionrate_france = np.gradient(val_france)
infectionrate_germany = np.gradient(val_germany)
#print(infectionrate_usa)


plt.plot(infectionrate_usa,label='USA',color='blue')
plt.plot(infectionrate_spain,label='Spain',color='red')
plt.plot(infectionrate_italy,label='Italy',color='green')
plt.plot(infectionrate_uk,label='UK',color='black')
plt.plot(infectionrate_france,label='France',color='violet')
plt.plot(infectionrate_germany,label='Germany',color='orange')
legend = plt.legend(loc='upper left', shadow=True, fontsize='medium')
plt.xlabel("Days")
plt.ylabel("Cases")
plt.title("Infection Rate")
plt.show()
```

```python
# Finding Maximum infection rate in each country

maxday_usa = np.argmax(infectionrate_usa)
print("USA had maximum infection rate on Day",maxday_usa)

maxday_spain = np.argmax(infectionrate_spain)
print("Spain had maximum infection rate on Day",maxday_spain)

maxday_italy = np.argmax(infectionrate_italy)
print("Italy had maximum infection rate on Day",maxday_italy)

maxday_uk = np.argmax(infectionrate_uk)
print("UK had maximum infection rate on Day",maxday_uk)

maxday_france = np.argmax(infectionrate_france)
print("France had maximum infection rate on Day",maxday_france)

maxday_germany = np.argmax(infectionrate_germany)
print("Germany had maximum infection rate on Day",maxday_germany)
```

```python
from numpy import array
from numpy.linalg import norm

#Computing L2 norm
norm1=norm(infectionrate_usa)
norm2=norm(infectionrate_spain)
norm3=norm(infectionrate_italy)
norm4=norm(infectionrate_uk)
norm5=norm(infectionrate_france)
norm6=norm(infectionrate_germany)

# Normalizing each vector by dividing infection rate vector with L2 norm values
normalusa=infectionrate_usa/norm1
normalspain=infectionrate_spain/norm2
normalitaly=infectionrate_italy/norm3
normaluk=infectionrate_uk/norm4
normalfrance=infectionrate_france/norm5
normalgermany=infectionrate_germany/norm6

#plotting normalized infection rate
plt.plot(normalusa,label='USA',color='blue')
plt.plot(normalspain,label='Spain',color='red')
plt.plot(normalitaly, label='Italy',color='green')
plt.plot(normaluk,label='UK',color='black')
plt.plot(normalfrance, label='France',color='violet')
plt.plot(normalgermany, label='Germany',color='orange')
legend = plt.legend(loc='upper left', shadow=False, fontsize='medium')
plt.xlabel("Days")
plt.ylabel("Infection")
plt.title("Normalized Infection Rate")
plt.show()
```
